```python
#vectorized functions

def prod_non_zero_diag_vect(X):
    diag = np.diag(X)
    return diag[diag != 0].prod()

def are_multisets_equal_vect(x, y):
    x_nums, x_counts = np.unique(x, return_counts=True)
    y_nums, y_counts = np.unique(y, return_counts=True)
    if(np.shape(x_nums) != np.shape(y_nums)):
        return False
    if(np.any(x_nums != y_nums) or np.any(x_counts != y_counts)):
        return False
    return True

def max_after_zero_vect(x):
    curr = np.ones((1))
    curr = np.hstack((curr, x))
    ind = curr == 0
    return np.max(x[ind[:-1]])

def convert_image_vect(img, coeff):
    return np.sum(img * coeff, axis=-1)

def run_length_encoding_vect(x):
    y = np.hstack((np.ones(1), x[:- 1]))
    first_positions = x != y
    first_positions[0] = True
    indexes_1 = np.arange(np.size(x))[first_positions]
    indexes_2 = np.hstack((indexes_1[1:], np.array([np.size(x)])))
    return x[first_positions], indexes_2 - indexes_1

def pairwise_distance_vect(X, Y):
    return np.sqrt(np.sum((X[:, np.newaxis] - Y) ** 2, axis=-1))
```

```python
#non vectorized functions
import math

def prod_non_zero_diag_non_vect(X):
    len_diag = min(len(X), len(X[0]))
    mult = 1
    for i in range(len_diag):
        if(X[i][i] != 0):
            mult *= X[i][i]
    return mult

def are_multisets_equal_non_vect(x, y):
    if(len(x) != len(y)):
        return False
    x.sort()
    y.sort()
    for i in range(len(x)):
        if(x[i] != y[i]):
            return False
    return True

def max_after_zero_non_vect(x):
    after_zero_list = list()
    for i in range(1, len(x)):
        if(not x[i - 1]):
            after_zero_list.append(x[i])
    return max(after_zero_list)

def convert_image_non_vect(img, coeff):
    height = len(img)
    width = len(img[0])
    result_img = list()
    for i in range(height):
        curr_str = list()
        for j in range(width):
            sum = 0
            for k in range(len(coeff)):
                sum += img[i][j][k] * coeff[k]
            curr_str.append(sum)
        result_img.append(curr_str)
    return result_img

def run_length_encoding_non_vect(x):
    val_list = [x[0]]
    num_list = list()
    counter = 1
    for i in range(1, len(x)):
        if(x[i - 1] == x[i]):
            counter = counter + 1
        else:
            val_list.append(x[i])
            num_list.append(counter)
            counter = 1
    num_list.append(counter)
    return (val_list, num_list)

def pairwise_distance_non_vect(X, Y):
    result_matr = list()
    for i in range(len(X)):
        curr_list = list()
        for j in range(len(Y)):
            dist = 0
            for k in range(len(X[0])):
                dist += (X[i][k] - Y[j][k]) ** 2
            curr_list.append(math.sqrt(dist))
        result_matr.append(curr_list)
    return result_matr
```

Точность предсказания немного улучшилась по сравнению с аналогичными моделями с параметром по умолчанию, но почти совпадает с качеством модели без использования регуляризации.

Постройте график зависимости $R^2$ для обучающей и тестовой (на кросс-валидации) выборок в зависимости от значения гиперпараметра. Для Lasso также постройте график зависимости количества ненулевых весов.

```python
lasso_r2_train = list()
lasso_r2_test = list()
lasso_coef = list()
ridge_r2_train = list()
ridge_r2_test = list()
for alpha in alphas_list:
    curr_lasso_r2_train = list()
    curr_lasso_r2_test = list()
    curr_ridge_r2_train= list()
    curr_ridge_r2_test= list()
    curr_lasso_coef = list()
    kf = KFold(n_splits=3, shuffle=False)
    for train_ind, test_ind in kf.split(y_train):
        lasso = Lasso(alpha=alpha).fit(x_train.iloc[train_ind], y_train[train_ind])
        ridge = Ridge(alpha=alpha).fit(x_train.iloc[train_ind], y_train[train_ind])
        curr_lasso_r2_test.append(lasso.score(x_train.iloc[test_ind], y_train[test_ind]))
        curr_lasso_r2_train.append(lasso.score(x_train.iloc[train_ind], y_train[train_ind]))
        curr_lasso_coef.append(np.sum(lasso.coef_ != 0))
        curr_ridge_r2_test.append(ridge.score(x_train.iloc[test_ind], y_train[test_ind]))
        curr_ridge_r2_train.append(ridge.score(x_train.iloc[train_ind], y_train[train_ind]))
    lasso_r2_test.append(mean(curr_lasso_r2_test))
    lasso_r2_train.append(mean(curr_lasso_r2_train))
    ridge_r2_test.append(mean(curr_ridge_r2_test))
    ridge_r2_train.append(mean(curr_ridge_r2_train))
    lasso_coef.append(curr_lasso_coef)                        
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))
ax1.semilogx(alphas_list, lasso_r2_test, label='test')
ax1.semilogx(alphas_list, lasso_r2_train, label='train')            
ax1.set_xlabel('значение параметра alpha')
ax1.set_ylabel('значение r2')
ax1.set_title('зависимость для метода с L1 регуляризацией')
ax1.grid(True)
ax1.legend()

ax2.semilogx(alphas_list, ridge_r2_test, label='test')
ax2.semilogx(alphas_list, ridge_r2_train, label='train')
ax2.set_xlabel('значение параметра alpha')
ax2.set_ylabel('значение r2')
ax2.set_title('зависимость для метода с L2 регуляризацией')
ax2.grid(True)
ax2.legend()

fig.show()
```

```python
plt.semilogx(alphas_list, lasso_coef)
plt.xlabel('значение параметра alpha')
plt.ylabel('количество ненулевых весов')
plt.title('зависимость для метода с L1 регуляризацией')
plt.grid(True)
```

В какой момент происходит недообучение? Почему?

__Ответ__: Можно заметить, что с уменьшением сложности модели (этому соответствует увеличение коэффициента регуляризации) точность резко снижается. В методе с L1 регуляризацией снижение точности начинается при alpha=0.1, при этом же значении некоторые веса становятся равными нулю, как видно из второго графика; при больших значениях параметра все веса обнуляются, точность модели так же стремится к нулю. В методе с L2 регуляризацией резкое снижение точности начинается при alpha=10. Также можно заметить, что оптимальным значением коэффициента регуляризации, подобранным по кросс-валидации, для обеих моделей является наименьшее из рассматриваемых, то есть лучшей признана самая сложная модель из рассматриваемых. Это тоже может свидетельствовать о том, что имеет место недообучение и увеличение сложности приведет к увеличению качества предсказания.
